```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
```

```python
NYC = pd.read_csv('../data/dataset_TSMC2014_NYC.csv')
TKY = pd.read_csv('../data/dataset_TSMC2014_TKY.csv')
```

```python
NYC.columns
```

```python
nyc_x = NYC[['latitude','longitude']].values
tky_x = TKY[['latitude','longitude']].values
```

```python
nyc_kmeans = KMeans(n_clusters=10,random_state=9).fit(nyc_x)
y_pred = nyc_kmeans.predict(nyc_x)
plt.scatter(nyc_x[:,0],nyc_x[:, 1],c=y_pred)
plt.show()
```

```python
nyc_kmeans.cluster_centers_
```

```python
plt.figure(figsize=(10,4))
visual_nyc1 = NYC[(NYC['latitude']>=40.9) &(NYC['latitude']<=41) &\
                 (NYC['longitude']>=-73.8) & (NYC['longitude']<=-73.7)]
visual_nyc1 = visual_nyc1.groupby(['venueCategory']).count()
visual_nyc1['userId'][10:40].plot.bar()
```

```python
plt.figure(figsize=(10,4))
visual_nyc1 = NYC[(NYC['latitude']>=40.6) &(NYC['latitude']<=40.7) &\
                 (NYC['longitude']>=-74.3) & (NYC['longitude']<=-74.0)]
visual_nyc1 = visual_nyc1.groupby(['venueCategory']).count()
visual_nyc1['userId'][10:40].plot.bar()
```

```python
tky_kmeans = KMeans(n_clusters=10,random_state=9).fit(tky_x)
y_pred_tky = tky_kmeans.predict(tky_x)
plt.scatter(tky_x[:,0],tky_x[:, 1],c=y_pred_tky)
plt.show()
```
